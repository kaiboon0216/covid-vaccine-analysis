# **What is the admitted and discharged rate of hospitals and pkrc in every state during this pandemic?**

```python
#create dataframes with columns needed from datasets
cases =state_cases[['date','state','cases_new']]
hosp =df_hosp[['date','state','admitted_total','discharged_total']]
```

```python
#merge datasets and turn the date column into year-month format for plotting used
hosp = hosp.merge(cases,left_on=['date','state'],right_on=['date','state'],how='left')
hosp['year_month'] = pd.to_datetime(hosp["date"], format='%Y-%m-%d').apply(lambda x: x.strftime('%Y-%m'))
```

```python
hosp_date =hosp[['year_month','admitted_total','discharged_total','cases_new']]
#group and sum data by month
hosp_date =hosp.groupby(['year_month'], as_index = False).sum()
#calculate admitted rate & discharged rate
hosp_date['Admitted rate'] = hosp_date['admitted_total']/hosp_date['cases_new']
hosp_date['Discharged rate'] = hosp_date['discharged_total']/hosp_date['admitted_total']
#get only admitted rate,discharged rate, and year_month column for plotting used
hosp_date = hosp_date[['year_month','Admitted rate','Discharged rate']]
```

```python
#melt data for plotting purpose
hosp_date = hosp_date.melt('year_month', var_name='type',  value_name='vals')
```

```python
#plot a grouped bar chart
chart=alt.Chart(hosp_date).mark_bar().encode(
    x='type:O',
    y='vals:Q',
    color='type:N',
)

#mark the value on each bar of the chart
text = chart.mark_text(
    align='center',
    baseline='middle',
    dy=-5  
).encode(
    text=alt.Text('vals', format='.1f')
)
alt.layer(chart, text, data=hosp_date).facet(column='year_month:N',title='Admitted and Discharged rate of hospital'
)
```

Based on the grouped bar chart above, the first few months of the Covid-19 pandemic which is from March until September of 2020, the Admitted rate of hospital is higher than the discharged rate of hospital.After September of 2020, the Admitted rate had keep decreasing until now only have slightly increase in the early of the 2021. Besides, the discharged rate of hospital did not have a extreme up and down, its rate only in between 0.6 to 1.2.  

```python
#group and sum data by month and state of the Malaysia
hosp_try=hosp.groupby(['year_month','state']).sum()
```

```python
hosp_try[hosp_try == 0] = 1
#calculate admitted rate & discharged rate
hosp_try['admitted_rate'] = hosp_try['admitted_total']/hosp_try['cases_new']
hosp_try['discharged_rate'] = hosp_try['discharged_total']/hosp_try['admitted_total']
hosp_try = hosp_try.reset_index()
#get only admitted rate,discharged rate, state, and year_month column for plotting used
hosp_try = hosp_try[['year_month','state','admitted_rate','discharged_rate']]
```

```python
#plot a line chart for admitted rate in every state of Malaysia
alt.Chart(hosp_try,title='Admitted rate of hospital in every state').mark_line().encode(
    x="year_month",
    y="admitted_rate",
    color = alt.Color('state', scale=alt.Scale(scheme = 'category20'))
).properties(
    width=1000,
    height=300
)
```

Based on the line chart above, we had observed that the hospital admitted rate in every state of Malaysia were very high at the beginning of the Covid-19 pandemic.Then, all the hospital admitted rate of the states were decrease rapidly in 2021 and only increase slighly during May to July of 2021 in Kelantan.

```python
#plot a line chart for discharged rate in every state of Malaysia
alt.Chart(hosp_try,title='Discharged rate of hospital in every state').mark_line().encode(
    x="year_month",
    y="discharged_rate",
    color = alt.Color('state', scale=alt.Scale(scheme = 'category20'))
).properties(
    width=1000,
    height=300
)
```

Based on the line chart above, we had observed that the hospital discharged rate in every state of Malaysia were not so high in the whole covid-19 pandemic.Only Negeri Sembilan had avery high discharged rate which was between 8 to 9 in July of 2020 and Pahang had 5.0 of discharged rate in August of 2020. The discharged rate of 2021 did not have a very high up and down, only between 0 to 2 of discharged rate.

# **Has vaccination helped to reduce the cases? Which states have shown the effectiveness of vaccination?**

```python
total_vac
```

```python
#get the population of Malaysia 
totalpop =df_static['pop'].head(1).values[0]
#calculate the percentage of population with second, first dose, and unvaccinated.
seconddose =vax_msia['cumul_full'].tail(1).values[0]/totalpop*100
firstdose = vax_msia['cumul_partial'].tail(1).values[0]/totalpop*100 -seconddose
total =(totalpop/totalpop)*100-firstdose-seconddose
```

```python
#create a dataframe with the data calculated
data = {'Type': [ 'Second Dose','First Dose','Unvaccinated'], 'Total': [seconddose,firstdose,total]}  
sum_df =pd.DataFrame(data = data,columns = ['Type','Total'])
#melt the data for plotting purpose
sum_df = sum_df.melt('Type', var_name='type',  value_name='vals')
sum_df
```

```python
#plot a stacked bar chart
chart = alt.Chart(sum_df,title='Number of citizens with First Dose/ Second Dose vaccine').mark_bar().encode(
    x='vals',
    y='type',
    color= alt.Color('Type', scale=alt.Scale(scheme = 'blues')),
    order=alt.Order(
      'type',
      sort='ascending'
    )
).properties(
    width=700,
    height=150
)
#add data value in every stack
text = alt.Chart(sum_df).mark_text(dx=-20, dy=1,color='black').encode(
    x=alt.X('vals', stack='zero'),
    y=alt.Y('type'),
    text=alt.Text('vals', format='.2f')
)

chart + text
```

Based on the stacked bar chart above, we can observed that there are near to 80% of population in Malaysia have been start their vaccination which is at least taking the first dose of the vaccine.Besides, there are  73.62% of population in Malaysia had been took the second dose of vaccine.

```python
#create a dataframe and add the needed column
Dose = vax_msia[['year_month','daily_partial','daily_full']]
#group and sum by year month 
Dose=Dose.groupby(['year_month'], as_index = False).sum()
Dose.rename(columns = {'daily_partial':'Partial Vaccination','daily_full':'Full Vaccination'}, inplace = True)
#melt the data for plotting use
Dose = Dose.melt('year_month', var_name='type',  value_name='vals')
#create a grouped bar chart
chart=alt.Chart(Dose).mark_bar().encode(
    x='type:O',
    y='vals:Q',
    color='type:N',
)
#add the value in each bar
text = chart.mark_text(
    align='center',
    baseline='middle',
    dy=-5  
).encode(
    text=alt.Text('vals', format='.0f')
)
alt.layer(chart, text, data=Dose).facet(column='year_month:N',title='Vaccination progress in Malaysia by month'
)
```

```python
#create dataframe and save the column needed from dataset
vax_type =vax_msia[['year_month','pfizer1','sinovac1','astra1','cansino']]
#rename column
vax_type.rename(columns = {'pfizer1':'Pfizer','sinovac1':'Sinovac','astra1':'AZ','cansino':'Cansino'}, inplace = True)
```

```python
#melt data for plotting purpose
month_type = vax_type.melt('year_month', var_name='type',  value_name='vals')
#group and sum by type and month
month_type=month_type.groupby(['year_month','type'], as_index = False).sum()
#create a line chart
alt.Chart(month_type,title='Vaccine reception of every types of vaccine in every month').mark_line().encode(
    x="year_month",
    y="vals",
    color = alt.Color('type', scale=alt.Scale(scheme = 'category20'))
).properties(
    width=1000,
    height=300
)
```

```python
#rename column
vax_type.rename(columns = {'pfizer1':'Pfizer','sinovac1':'Sinovac','astra1':'AZ','cansino':'Cansino'}, inplace = True)
#group and sum by month
vax_type=vax_type.groupby(['year_month'], as_index = False).sum()
#calculate the percentage of every types vaccine in every month
vax_type['Total(%)']= vax_type['Cansino']+vax_type['AZ']+vax_type['Sinovac']+vax_type['Pfizer']
vax_type['Cansino(%)'] =vax_type['Cansino']/vax_type['Total(%)']*100
vax_type['AZ(%)'] =vax_type['AZ']/vax_type['Total(%)']*100
vax_type['Sinovac(%)'] =vax_type['Sinovac']/vax_type['Total(%)']*100
vax_type['Pfizer(%)'] =vax_type['Pfizer']/vax_type['Total(%)']*100
#get the columns needed
vax_type =vax_type[['year_month','Cansino(%)','AZ(%)','Sinovac(%)','Pfizer(%)']]
#melt data for plotting purpose
vax_type = vax_type.melt('year_month', var_name='type',  value_name='vals')
vax_type =vax_type[vax_type!=0]
```

```python
#create a stacked bar chart
chart=alt.Chart(vax_type,title='Percentage of Vaccine reception in every types of vaccine').mark_bar().encode(
    x='year_month',
    y='vals',
    color= alt.Color('type', scale=alt.Scale(scheme = 'set2')),
    order=alt.Order(
      'type',
      sort='ascending'
    )
).properties(
    width=800,
    height=500
)

chart 
```

Based on the graph above, Pfizer has been kept receiving by the Malaysia citizen from the beginning of vaccination start until now. Sinovac has start provide from April 2021 and has the highest amount of receive in July which is about 4500K of receive, but decrease extremely in September 2021.AZ also almost same with the Sinovac which is start in May and keep decreasing until September 2021. Only the Cansino has been started recently which is August 2021.
